```python
import pandas as pd
```

```python
EXPERIMENTS = 10
REPORTS_PATH = '../data/reports/'
```

```python
reports = []
for i in range(EXPERIMENTS):
    file = REPORTS_PATH + f'comparison_experiment_{i + 1}.csv'
    temp_df = pd.read_csv(file)
    reports.append(temp_df)

df = pd.concat(reports)
```

```python
df = df.reset_index().drop(columns='Unnamed: 0').rename(
    columns={'index': 'Range_ID'}).sort_values(by='Composite Score', ascending=False, ignore_index=True)
```

```python
df
```

```python
df.to_csv('../data/reports/comparison_experiments.csv', mode='a', header=False)
```

### **Analysing the Best Subset ID Range**
